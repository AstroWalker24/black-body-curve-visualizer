```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.constants import c,h,k,Wien
import ipywidgets as widgets 
print(c,h,k)
```

```python


# calculating the spectral radiance of the blackbody 
def spectral_radiance(Temperature,wavelength):


    # calculate the spectral radiance 
    exponent = (h * c) / (wavelength * k * Temperature)
    denominator = np.exp(exponent) - 1
    
    return (2 * h * c**2) / (wavelength**5 * denominator)



```

```python
wavelengths= np.linspace(1e-9,3e-6,1000)
sun_temp = 5778

spectral_radiance_sun=spectral_radiance(sun_temp,wavelengths)
# print(spectral_radiance_sun)


plt.plot(wavelengths,spectral_radiance(sun_temp,wavelengths))
plt.xlabel('Wavelength (nm)')
plt.ylabel('Spectral Radiance ')
plt.title('Spectral Radiance of the Sun')
plt.show()

```

```python
# using the interactive widget to change the temperature of the blackbody
def plot_blackbody_radiance(Temperature):
    plt.plot(wavelengths,spectral_radiance(Temperature,wavelengths))
    plt.xlabel('Wavelength (nm)')
    plt.ylabel('Spectral Radiance ')
    plt.title('Spectral Radiance of the Blackbody')
    plt.show()

widgets.interact(plot_blackbody_radiance,Temperature=(1000,10000,1000))
```
